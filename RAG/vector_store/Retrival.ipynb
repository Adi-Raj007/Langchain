```python
docs_path=("/home/aditya/Desktop/CAT-2024-Slot-01-Final-with-Answer-Keys.pdf")
loader=PyPDFLoader(docs_path)
print(docs_path)
docs=loader.load()
# print(docs[1].page_content)
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="/home/aditya/Desktop/PRO/model/jina-embeddings-v2-base-en",
    model_kwargs={"device": "cpu"},
    encode_kwargs={"normalize_embeddings": True}
)
```

```python
vector_space = Chroma.from_documents( 
    documents=docs,
    embedding= embedding_model,
    persist_directory='chroma_data',
    collection_name='sample'

)
```

```python
vector_space.get(include=["embeddings","metadatas","documents"])
```

```python
filtered_doc= vector_space.similarity_search(
    query=" Which one of the following statements provides a gist of this passage ?",
    k=2
)
```

```python
for i, doc in enumerate(filtered_doc):
    print(f"\n--- Result {i+1} ---")
    print(f"Content:\n{doc.page_content}...")  
```
